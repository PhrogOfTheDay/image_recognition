```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
training_df = pd.read_csv('train.csv')
testing_df = pd.read_csv('test.csv')
```

## Preprocessing

```python
training_df.isnull().sum() #no nulls
```

```python
testing_df.isnull().sum() 
```

```python
#no nulls, and the nature of the dataset makes usual preprocessing like anomaly detection, fillna, feature encoding unnecessary
```

## Processing the Data

```python
model = LogisticRegression(max_iter=1000)
X_train , y_train = training_df.drop(['label'], axis=1), training_df[['label']]
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(testing_df)
for x in range(20):
    modified = testing_df.iloc[x,:]
    fig, (ax_plot, ax_text) = plt.subplots(1, 2, figsize=(10, 4))
    ax_plot.matshow(modified.to_numpy().reshape(28,28))
    ax_text.text(0, 1, predictions[x], va='center', ha='center', fontsize=20, family='monospace')
    ax_text.axis('off')
    plt.tight_layout()
```
